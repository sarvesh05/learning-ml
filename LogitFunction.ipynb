```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def logit(x_):
    #x_ = np.arange(-5,5)
    return list(map(lambda x : 1 / (1 + np.e**-x), x_))

x_ = np.arange(-5,6)
y = logit(x_)

plt.plot(x_, y, color = 'blue', marker = "*")
plt.title("Logit function") 
plt.xlabel("Input") 
plt.ylabel("Output") 
plt.show()
```
